```python
from sklearn.datasets import load_iris
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Загрузка данных
data = load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target
```

```python
print(df.isnull().sum())
```

Пропусков нет

Построим корреляционную матрицу:

```python
correlation = df.corr(numeric_only=True)
print(correlation)

# Построение тепловой карты
plt.figure(figsize=(10, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Корреляционная матрица признаков')
plt.show()
```

Корреляционный анализ показал, что наиболее сильные положительные корреляции наблюдаются между признаками petal length (cm) и petal width (cm) (0.96), а также между этими признаками и целевой переменной target (0.95 и 0.96 соответственно). Это свидетельствует о высокой информативности этих признаков для построения моделей машинного обучения. В то же время признаки sepal width (cm) и target имеют отрицательную слабую корреляцию (-0.42), что говорит о низкой значимости этого признака для классификации. Таким образом, для построения моделей стоит в первую очередь использовать признаки, связанные с лепестками, так как они вносят наибольший вклад в предсказание класса.

"Парные диаграммы"

```python
sns.pairplot(df, hue='target', palette='husl')
plt.suptitle("Парные диаграммы признаков набора Iris", y=1.02)
plt.show()
```
